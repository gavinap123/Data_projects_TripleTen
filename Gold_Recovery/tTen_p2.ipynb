# 3. Build the model

```python
# 3.1. Function to calculate sMAPE
def smape(target, prediction):
    target = np.array(target)
    prediction = np.array(prediction)

    denominator = np.abs(target) + np.abs(prediction) / 2
    smape_value = 100 * np.mean(np.abs(target - prediction) / denominator)
    
    return smape_value
```
